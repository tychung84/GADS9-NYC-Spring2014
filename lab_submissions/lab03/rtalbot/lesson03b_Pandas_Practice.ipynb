```python
for x in iris.Name.unique():
    print 'Flower Name: %s' % x
    print 'Max Sepal Length: %.2f' % iris[(iris.Name == x)].SepalLength.max()
    print 'Mean Sepal Length: %.2f' % iris[(iris.Name == x)].SepalLength.mean()
    print 'Min Sepal Length: %.2f' % iris[(iris.Name == x)].SepalLength.min()
    print 'Median Sepal Length: %.2f' % iris[(iris.Name == x)].SepalLength.median()
```

```python
def FlowerStats(flower):
    for feature in iris.columns.values:
        if iris[feature].dtypes != object:
            print { 
                    'Flower' : flower,
                    'Feature' : feature,
                    'Max' : iris[(iris.Name == flower)][feature].max(),
                    'Min' : iris[(iris.Name == flower)][feature].min(),
                    'Mean' : iris[(iris.Name == flower)][feature].mean(),
                    'Median' : iris[(iris.Name == flower)][feature].median()
                    }
        
for name in iris.Name.unique():
    FlowerStats(name)
```
