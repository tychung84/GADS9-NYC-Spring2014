```python
%pylab inline

import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
```

```python
lem = pd.read_csv('data/lemon_training.csv', index_col='RefId')
```

```python
lem.head(10)
```

```python
lem.describe()
```

```python
lem.info()
```

```python
print lem.groupby('Make').mean()[['IsBadBuy']]
```

```python
#how many miles were put on this car per year?
lem['MilesPerYear'] = lem.VehOdo / (lem.VehicleAge+1)
lem.MilesPerYear.hist(bins=30)

#clean long-tail distribution with hump on left-side
```

```python
#center the data
np.log(lem.MilesPerYear).hist(bins=30)
```

```python
X_makes = pd.get_dummies(lem['Make'])
X_joined = pd.DataFrame(lem['MilesPerYear'])
X_joined = X_joined.join(X_makes)
X_joined.info()
```

```python
Y = lem['IsBadBuy']
X_train, X_test, Y_train, Y_test = train_test_split(X_joined,Y,test_size=0.25)

clf = RandomForestClassifier()
clf.fit(X_train, Y_train)
score = clf.score(X_test, Y_test)
print "Random Forest %0.2f%%" % (score * 100.0)
```

```python
lem_test = pd.read_csv('data/lemon_test.csv', index_col='RefId')
```

```python
lem_test.info()
```

```python
lem_test['MilesPerYear'] = lem_test.VehOdo / (lem_test.VehicleAge+1)
```

```python
test_X_makes = pd.get_dummies(lem_test['Make'])
test_X_joined = pd.DataFrame(lem_test['MilesPerYear'])
test_X_joined = test_X_joined.join(test_X_makes)
test_X_joined.info()
```

```python
y_pred = clf.predict(test_X_joined)
```

```python
# Create a submission
submission = pd.DataFrame({ 'RefId' : test_X_joined.index.values, 'prediction' : y_pred })
submission.describe()
```
