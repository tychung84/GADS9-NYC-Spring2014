During the course of the campaign, the field staff did basic polls to determine who did/didn't support the candidate. These results are found in the **support_level** column.  Those who were not polled appear as **NaN** in the imported dataset.  To correct this, I replaced all the **NaN**'s with 0's to indicate that we did not have a support score for those individuals.  

To determine which features could be used across the training set and the final set I decided to do counts for each subset of the data and remove the features that had the least observations in both subsets.

To make the most out of the information I have when creating the model, I decided to make new boolean columns based on information that is available.  For example, creating 0 or 1 flags for people who we have cell number for is much more useful than actually knowing the cell numbers.

```python
#Created a function that checks to see if the length of a number is less than or greater than 5.
def mobile(x):
    if len(str(x)) < 5:
        return 0
    else:
        return 1
    
df['mobileflag'] = df['mobile_number'].apply(mobile)   
df['phoneflag'] = df['phone_number'].apply(mobile) 
df['emailflag'] = df['email1'].apply(mobile)
df['addressflag'] = df['address_zip'].apply(mobile)


def district(x):
    if len(str(x)) == 0:
        return 0
    else:
        return 1

df['districtflag'] = df['county_district'].apply(district)  
df['villageflag'] = df['village_district'].apply(district)


def party(x):
    if (x) == 'D' or 'G':
        return 1
    else:
        return 0
    
df['partyflag'] = df['party'].apply(party)  



def sexflag(x):
    if (x) == 'F':
        return 1
    if (x)== 'M':
        return 2
    else:
        return 0
    
df['sexflag'] = df['sex'].apply(sexflag)  
```
