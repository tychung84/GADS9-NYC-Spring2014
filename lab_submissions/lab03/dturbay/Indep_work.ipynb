```python
def short_or_long(x): 
 return 'long' if x > 4 else 'short'

iris['PetalLengthSummary'] = iris.PetalLength.apply(short_or_long)

iris.head()
```

```python
fun = lambda x: x < iris.x.mean()
def levelPL(x):
    if x < iris.PetalLength.mean():
        return 'below average'
    elif x > iris.PetalLength.mean():
        return 'above average'
    else:
        return 'exactly average'
    
def levelSL(x):
    if x < iris.SepalLength.mean():
        return 'below average'
    elif x > iris.SepalLength.mean():
        return 'above average'
    else:
        return 'exactly average'
```

```python
iris['PL:over/underAvg'] = iris.PetalLength.apply(levelPL)

iris['SL:over/underAvg'] = iris.SepalLength.apply(levelSL)


iris
```
