```python
#Create a data frame with smaller features from the original much larger data frame
bbdf = pd.DataFrame(bbdata, columns=['Date','Day','Visitors','VGameNumber','Home','HGameNumber','VScore','HScore','DayNight','ParkID','Attendance','VHRs','HHRs'])
```

```python
#Add feature columns
#bbdf['GameScore'] = bbdf.VScore + bbdf.HScore
#bbdf['HRs'] = bbdf.VHRs + bbdf.HHRs
bbdf['Year'] = [t[0:4] for t in bbdf.Date]
bbdf['Month'] = [t[4:6] for t in bbdf.Date]
#bbdf['Day2'] = [t[6:8] for t in bbdf.Date]
bbdf['Winner'] = [bbdf.Visitors[i] if bbdf.VScore[i]>bbdf.HScore[i] else bbdf.Home[i] for i in range(0,len(bbdf))]
bbdf['NightGame'] = [1 if bbdf.DayNight[i]=='N' else 0 for i in range(0,len(bbdf))]
```

```python
len(bbdf[bbdf.Attendance == 0])
bbdf[bbdf.Attendance==0]
```

```python
#There 164 games with zero attendance which is approx. 10% of the dataset.
# I could substitute in an average attendance for that ParkID, but for now I will not use those rows.

bbdf.Attendance[bbdf.Attendance==0] = np.nan
```

```python
bbdf['PercAttn'] = [bbdf.Attendance[i] / (bbdf['Attendance'].groupby(bbdf.ParkID==bbdf.ParkID[i]).max()) for i in range(0,len(bbdf))]
```

```python
bbdf['Attendance'].groupby(bbdf.ParkID=='BOS07').max()
```

```python
bbdf['PercAttn'] = [float(str(bbdf['PercAttn'][i]).split()[4]) for i in range(0,len(bbdf))]
```

```python
bbdf['IsProfit'] = [1 if bbdf.PercAttn[i]>.9 else 0 for i in range(0,len(bbdf))]
```
