```python
#Create a data frame with smaller features from the original much larger data frame
bbdf = pd.DataFrame(bbdata, columns=['Date','Day','Visitors','VLeague','Home','HLeague','VScore','HScore','DayNight','ParkID','Attendance','VHRs','HHRs'])
```

```python
#Add feature columns
bbdf['GameScore'] = bbdf.VScore + bbdf.HScore
bbdf['HRs'] = bbdf.VHRs + bbdf.HHRs
bbdf['Year'] = [t[0:4] for t in bbdf.Date]
```

```python
# Remove any zero attended games
print len(bbdf[bbdf.Attendance == 0])

#probably best to flag those bad data as NAN
bbdf.Attendance[bbdf.Attendance==0] = np.nan
```
