```python
def PLshort_or_long(x):
    return 'long' if x > data.PetalLength.mean() else 'short'

data['PetalLengthSummary'] = data.PetalLength.apply(PLshort_or_long)
data[55:61]
```

```python
def PWshort_or_long(x):
    return 'long' if x > data.PetalWidth.mean() else 'short'

data['PetalWidthSummary'] = data.PetalWidth.apply(PWshort_or_long)
```

```python
def SLshort_or_long(x):
    return 'long' if x > data.SepalLength.mean() else 'short'

data['SepalLengthSummary'] = data.SepalLength.apply(SLshort_or_long)
```

```python
def SWshort_or_long(x):
    return 'long' if x > data.SepalWidth.mean() else 'short'

data['SepalWidthSummary'] = data.SepalWidth.apply(SWshort_or_long)
```
