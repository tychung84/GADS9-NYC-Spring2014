```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#GDP data by Country and year
gdp = pd.read_csv('gdp.csv', sep='\t')
#Region info by country
ind = pd.read_csv('indicator.csv')

print len(gdp)
print len(ind)
```

```python
#make the index the country code
gdp.index = gdp['Country Code']
ind.index = ind['Country Code']
```

```python
#remove extra column for unclassified GDP (we wont need it)
gdp = gdp[gdp['Country Code'] != 'INX']
len(gdp)

#want to add region from ind as a column to gdp
for cc in gdp['Country Code']:
        gdp.loc[cc,'Region'] = ind.loc[cc,'Region']

gdp['Region'].head()
```

```python
#average gdp by region
region_agdp = gdp.groupby('Region').mean()
```

```python
%pylab inline


region_agdp.T.plot(figsize=(10,5),title='Average GDP by Region',logy=False)
#try log scale
region_agdp.T.plot(figsize=(15,8),title='Average GDP by Region',logy=True)
#North America seems to be throwing things off...
region_agdp[region_agdp.index != 'North America'].T.plot(figsize=(10,5),title='Average GDP by Region',logy=False,ylim=(0,700000000000),legend=True)
```

```python
region_agdp[region_agdp.index != 'North America'].plot(figsize=(13,8),title='Average GDP by Region',logy=False,ylim=(0,700000000000),kind='bar',legend=False)
```

```python
a = region_agdp[region_agdp.index != 'North America'].T

fig, ax = plt.subplots(nrows=1, ncols=3,figsize=(15, 5))

ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)
```

```python
latam = region_agdp[region_agdp.index == 'Latin America & Caribbean'].T
col = gdp.loc['COL'][4:-1].dropna()
bra = gdp.loc['BRA'][4:-1].dropna()
arg = gdp.loc['ARG'][4:-1].dropna()
```

```python
latam.plot(logy=True, figsize=(10,5),style='--')
bra.plot()
col.plot()
arg.plot()
```
