```python
xtrain, xtest, ytrain, ytest = train_test_split(lemon_train[['VehOdo', 'VehicleAge', 'VehBCost']], lemon_train.IsBadBuy)
```

```python
clf = MultinomialNB().fit(xtrain, ytrain)

print "Accuracy: %0.2f%%" % (100 * clf.score(xtest, ytest))
```

```python
bad_buy_predictor = clf.predict(xtest)
```
