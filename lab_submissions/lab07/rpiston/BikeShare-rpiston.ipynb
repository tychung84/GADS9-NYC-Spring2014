```python
%pylab inline
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
day_data = pd.read_csv('./data/day.csv')
day_data.head()
```

##Clean

```python
day_data['temp'] = day_data['temp']*41
day_data['atemp'] = day_data['atemp']*50
day_data['hum'] = day_data['hum']*100
day_data['windspeed'] = day_data['windspeed']*67
day_data.head()
```

```python
day_data.temp.describe()
fig = plt.figure()
fig.set_figwidth(20.0)

ax1 = fig.add_subplot(1,3,1)
ax1.scatter(day_data['atemp'], day_data['cnt'])

ax2 = fig.add_subplot(1,3,2)
ax2.scatter(day_data['atemp'], day_data['registered'])

ax3 = fig.add_subplot(1,3,3)
ax3.scatter(day_data['atemp'], day_data['casual'])

print fig
```

```python
lin_f, lin_p = f_regression(day_data[['atemp']].values, day_data['cnt'].values)
print 'LINEAR F-Test Values:', lin_f[0]
print 'LINEAR p-values:     ', lin_p[0]
```

```python
lin_f, lin_p = f_regression(day_data[['temp']].values, day_data['cnt'].values)
print 'LINEAR F-Test Values:', lin_f[0]
print 'LINEAR p-values:     ', lin_p[0]
```

```python
lin_f, lin_p = f_regression(day_data[['season']].values, day_data['cnt'].values)
print 'LINEAR F-Test Values:', lin_f[0]
print 'LINEAR p-values:     ', lin_p[0]
```

```python
linear_fit = LinearRegression()

x1 = day_data[['atemp']]
y1 = day_data['casual']
linear_fit.fit(x1,y1)
model = linear_fit.predict(x1)

print "sklearn intercept and coef (linear):", linear_fit.intercept_, linear_fit.coef_

plt.scatter(x1,y1)
plt.plot(x1,model)
plt.ylabel('casual users')
plt.xlabel('feel temperature')
plt.title('Number of Casual Users')
```

```python
print "Linear R-squared", round(linear_fit.score(day_data[['atemp']].values, day_data['casual'].values), 4)
```
