```python
import pandas as pd
import numpy as np

iris = pd.read_csv('https://raw.github.com/pydata/pandas/master/pandas/tests/data/iris.csv')
iris
```

```python
#1.How many different names of flowers are in this data set?

iris2 = iris.groupby(['Name']).sum() 

iris2
#This will group the entire set of data by the last column "Name".
#Gives 3 groups and the sums of the other.
```

```python
#2.How many exist of each type of flower?
iris.groupby(['Name']).count()
#Indicates there is 50 of each flower type. 
```

```python
#3.Determine the min, median, mean, max for each numeric feature in the data set.
iris.describe()
```

```python
#4.Determine the same for each flower type.

iris3 = iris.groupby(['Name']).describe()
iris3

#This should spit out the analysis but based on flower type groups. 
```

```python
#5a. How does the shape of these results change from the average of all flowers? 

iris.groupby(['Name']).mean()
#Looks like sepal length, petal length and petal width increases: Setosa<Versicolor<Virginica
#Looks like sepal width deeases: Setosa>Versicolor>Virginica
```
