### Utility Functions for loading, processing, munging, removing stop words.

```python
stop = stopwords.words('english')
def is_stop_word(word):
    if len(word)<=2: return True
    if contains_digits(word): return True               
    if word.lower() in stop:return True
    else:return False
```

```python
porter = PorterStemmer()
wl = WordNetLemmatizer()
```

```python
def munge_word(word):
    munge_list = [ ".", ":", "", "?", ",", "'","(", ")", "),","[", "]" , "\"", "-", "&", "e.g", "E.g"]
    for l in munge_list:
        if l in word:
            word = word.strip(l)         
    word = unicode(word, 'ascii', 'ignore')
    word = porter.stem(word)
    #word = wl.lemmatize(word)
    return word
```

```python
_digits = re.compile('\d')
def contains_digits(d):
    return bool(_digits.search(d))
```

```python
def load_n_process_data( data_dir, word_limit= 7000, debug=False ):
    
    dance_train_data = {}
    
    for dance_file in os.listdir( data_dir ):
        dance_name_label = str.split(dance_file,".")[0]
        dance_train_list = []
        f = open(data_dir + dance_file, 'r')
        if debug == True:
            print "Reading file %s" %f
        
        i = 0
        for line in f:
            for word in line.split():
                i+=1
                if i == word_limit: break
                clean_word = munge_word(word)
               
                if clean_word != "":
                    if not is_stop_word(clean_word):
                        dance_train_list.append (clean_word)
            if i == word_limit: break
        
        all_words_freq = nltk.FreqDist(w.lower() for w in dance_train_list)
        dance_train_data [dance_name_label] = all_words_freq

        #word_features = all_words_freq.keys()[:feature_limit]
        #word_features = all_words_freq.keys()
        #dance_train_data [ dance_name_label ] = word_features
        f.close()
        
    if debug == True:
        for key, value in dance_train_data.items():
            print "--------------------------------------------------------------------"
            print (key, len(value))
            print value
        
    return dance_train_data
```

```python
def process_features_n_words(dance_data_dict, feature_limit=100, debug=False):
    temp_list = []    
    dance_unique_word_dict = {}
    
    # Remove common words
    for data in dance_data_dict.values():
        l = data.keys()
        temp_list.append(l)
    
    counts = Counter(chain(*map(set,temp_list)))
    temp_list = [[i for i in sub_list if counts[i]==1] for sub_list in temp_list]
    
    i=0
    for key in dance_data_dict.keys():
        dance_unique_word_dict[key] = temp_list[i]
        i+=1
   
    # Keep the top most occuring features in the unique word dictionary
    for dance_name_label, word_list in dance_unique_word_dict.items():
        all_words_freq = nltk.FreqDist(w.lower() for w in word_list)
        word_features = all_words_freq.keys()[:feature_limit]
        dance_unique_word_dict [ dance_name_label ] = word_features
    
    if debug == True:
        print "---------------------------------------------------------------"
        print "    Processed features and words dictionary                    "
        print "---------------------------------------------------------------"
        for key, value in dance_unique_word_dict.items():
            print (key, len(value))
            print value
    
    return dance_unique_word_dict 
```

### TF-IDF Classifier : Calculating the global frequency count of words across all documents

```python
def get_global_freq_count(dance_data_dict, debug=False):
    global_term_freq = {}

    # Compute the Global Term Frequency
    for global_terms_in_doc in dance_data_dict.values():
        for (word,freq) in global_terms_in_doc.items():
            if word in global_term_freq:
                global_term_freq[word] += 1
            else:
                global_term_freq[word] = 1
            
    if debug == True:
        print "---------------------------------------------------"
        print "Global Term Frequency"
        print "---------------------------------------------------"
        print global_term_freq
        print "___________________________________________________"
    
    return global_term_freq
```

### TF-IDF Classifier : TF-IDF Algorithm Implementation

```python
def compute_tfidf(dance_classify_data, dance_train_data, debug=False):
    tfidf_dict = {}
    
    num_docs = len(dance_train_data.keys())
    if debug == True:
        print "--------------------------------------------------------------------"
        print "Total Number of docs applied in classification = %d" %num_docs
        print "--------------------------------------------------------------------"

    global_term_freq = get_global_freq_count(dance_train_data)
    
    columns = dance_train_data.keys()
    indexes = dance_classify_data.keys()
      
    tfidf_result_array = np.zeros([len(indexes), len(columns)], dtype=np.float64)
                                   
    # Compute the TF-IDF for the words
    i=0
    for c_term in dance_classify_data.keys():
        tfidf_result_row = []
        
        # For each word, compute the TF_IDF value in each document, iterating through the documents
        for dance_train_word_dict in dance_train_data.values():
            max_freq = 0;
            
            if c_term in dance_train_word_dict.keys():
                freq = dance_train_word_dict[c_term]
            else:
                freq = 0
                
            max_freq = max(dance_train_word_dict.values())        
                    
            if c_term in global_term_freq.keys():
                c_term_global_freq = global_term_freq[c_term]
            else:
                c_term_global_freq = 0
                    
            idf = math.log(float(1 + num_docs) / float(1 + c_term_global_freq))
            
            #augmented frequency, to prevent a bias towards longer documents, 
            #e.g. raw frequency divided by the maximum raw frequency of any term in the document:
            tfidf = float(freq) / float(max_freq) * float(idf)
            
            tfidf_result_row.append(tfidf)

        tfidf_result_array[i] = np.array([tfidf_result_row])
        i+=1
    
    df = pd.DataFrame(tfidf_result_array, index=indexes, columns=columns)
        
    if debug == True:
        print "----------------------------------------------"
        print "Printing TF-IDF DataFrame Values"
        print "----------------------------------------------"
        print  df.describe()
        print "----------------------------------------------"
        print df
        print "----------------------------------------------"

    return df
```

### TF-IDF Classifier : A new classifier that classifies text using TF-IDF algorithm!                     Lets see how this works?!!!

```python
def classify_with_tfidf(df, debug=False):
    
    column_sum_list = list(df.sum(axis=0))
    
    if debug == True:
        print "TF-IDF word sum in documents"
        print column_sum_list
    
    matched_doc = df.columns[column_sum_list.index(max(column_sum_list))]

    if debug == True:
        print "------------------------------------------------------"
        print "Printing TF-IDF DataFrame Values"
        print "------------------------------------------------------"
        print  df.describe()
        print "------------------------------------------------------"
        print df
        print "------------------------------------------------------"

    return matched_doc
```

```python
def show_most_informative_features(df, num_of_features):
    
    row_max_values = pd.DataFrame(df.max(axis=1), columns=['max_tfidf_score_across_docs'])
    print "\n\n+++++++++++++++++++++++++++++"
    print "   Most Informative Features"
    print "+++++++++++++++++++++++++++++"
    print row_max_values.sort(columns=['max_tfidf_score_across_docs'], ascending=False, axis=0).head(num_of_features)
```

### Create Bag of Words Features for Training and Test dance data sets

```python
def bag_of_words(words):
    return dict([(word, True) for word in words])
```

```python
def create_features(dance_dict):
    feature_list = []
    for key, value in dance_dict.items():
        feature_list += [(bag_of_words(value), key)]
    return feature_list
```
