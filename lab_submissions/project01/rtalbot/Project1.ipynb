```python
diamonds = pd.read_csv('data/diamonds.csv')
```

## Clean the Data

1. Give the cut a score (Fair, Good, Very Good, Premium, Ideal)
2. Give the color a score, J (worst), I, H, G, F, E, D (best)
3. Give the clarity a score, I1 (worst), SI1, SI2, VS1, VS2, VVS1, VVS2, IF (best)
4. Calculate an overall score (carat * cutScore * colorScore * clarityScore)

```python
diamonds.color.unique()
```

```python
colorScore = []
cutScore = []
clarityScore = []
colors = ['0','J','I','H','G','F','E','D']
cuts = ['0','Fair','Good','Very Good','Premium','Ideal']
clarity = ['0','I1','SI1','SI2','VS1','VS2','VVS1','VVS2','IF']

for m in diamonds.color:
        if m in colors:
            colorScore.append(colors.index(m))
        else: colorScore.append(0)

for m in diamonds.cut:
    if m in cuts:
        cutScore.append(cuts.index(m))
    else: cutScore.append(0)
        
for m in diamonds.clarity:
    if m in clarity:
        clarityScore.append(clarity.index(m))
    else: clarityScore.append(0)
        
diamonds['cutScore']=cutScore
diamonds['colorScore']=colorScore
diamonds['clarityScore']=clarityScore
diamonds.head()
```

```python
diamonds[['carat','price','colorScore','cutScore','clarityScore']].describe()
```

```python
print len(diamonds[diamonds.colorScore==0])
print len(diamonds[diamonds.cutScore==0])
print len(diamonds[diamonds.clarityScore==0])
```

```python
diamonds['OverallScore']=diamonds.cutScore*diamonds.clarityScore*diamonds.colorScore*diamonds.carat
diamonds.head()
```
