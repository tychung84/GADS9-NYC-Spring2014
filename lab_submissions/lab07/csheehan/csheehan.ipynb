### Check individual  p scores of each variable against casual, registered, and total riders

```python
plt_temp = plt.figure()
#this solves the Y axis units but messes up the spacing

#plt_temp, axes = plt.subplots(nrows=1, ncols=3)
# this solves the spacing but messes up the Y axis units
plt_temp.set_figwidth(4)
plt_temp.set_figheight(10)

ax1 = plt_temp.add_subplot(3,1,1)
ax2 = plt_temp.add_subplot(3,1,2)
ax3 = plt_temp.add_subplot(3,1,3)

#plt_temp.tight_layout()

fit_temp_cas = LinearRegression()
fit_temp_cas.fit(bike[['temp']].values, bike['casual'].values)
print  fit_temp_cas.intercept_, fit_temp_cas.coef_

cas_X = bike[['temp']].values
cas_Y = bike['casual'].values
predict_cas_temp = fit_temp_cas.predict(bike[['temp']].values)

ax1.scatter(cas_X,cas_Y)
ax1.plot(cas_X,predict_cas_temp, color='red')
ax1.set_ylabel('Total Riders')
ax1.set_xlabel('Temperature')

#-------------------------------------

fit_temp_reg = LinearRegression()
fit_temp_reg.fit(bike[['temp']].values, bike['registered'].values)
print  fit_temp_reg.intercept_, fit_temp_reg.coef_

cas_X = bike[['temp']].values
cas_Y = bike['registered'].values
predict_reg_temp = fit_temp_reg.predict(bike[['temp']].values)

ax2.scatter(cas_X,cas_Y)
ax2.plot(cas_X,predict_reg_temp, color='red')
ax2.set_ylabel('Registered Riders')
ax2.set_xlabel('Temperature')

#-------------------------------------

fit_temp_cnt = LinearRegression()
fit_temp_cnt.fit(bike[['temp']].values, bike['cnt'].values)
print  fit_temp_cnt.intercept_, fit_temp_cnt.coef_

cas_X = bike[['temp']].values
cas_Y = bike['cnt'].values
predict_cnt_temp = fit_temp_cnt.predict(bike[['temp']].values)

ax3.scatter(cas_X,cas_Y)
ax3.plot(cas_X,predict_cnt_temp, color='red')
ax3.set_ylabel('Total Riders')
ax3.set_xlabel('Temperature')

print plt_temp
```
