```python
# pull out the columns i want (Make, MilesPerYear, CostPerAnnualMile)
X_rec = lemon_train[['Make', 'MilesPerYear', 'CostPerAnnualMile']]
y = lemon_train['IsBadBuy']
```

```python
X_rec.info()
```

```python
makes = lemon_train['Make']
makes = pd.get_dummies(makes)
makes.tail()
```

```python
X_joined = X_rec.join(makes,how='left')
X_joined.tail()
```

```python
#Delete Make column from X_joined before passing to classifier
X_joined = X_joined.drop('Make',1)
```

```python
X_joined.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_joined,y, test_size=0.3)
#X_joined.head(), y.head()
```

```python
nb_clf = MultinomialNB().fit(X_train, y_train)
```

```python
X_train
```

```python
train_acc = nb_clf.score(X_train, y_train)
test_acc = nb_clf.score(X_test, y_test)

print "Training Acc: %0.2f" % (train_acc * 100)
print "Test Acc: %0.2f" % (test_acc * 100)
#this is pretty high
#if 12% were lemons, is this just saying they're all non-lemons and taking the 12% hit? -> looking at submission.csv, yes
```

```python
train_predict = nb_clf.predict(X_train)
```
